В забеге удачно преодолели дистанцию (91,7%;91,8%) участников. 
Каково минимальное число участников?

```python
from sympy.abc import x, n
from sympy import Eq
from sympy import solve

expr1 = x >= 0.917 * n
expr2 = x <= 0.918 * n
syseq = [Eq(n, 1000), Eq(n, 500), Eq(n, 200), 
         Eq(n,100), Eq(n, 73)]
for eq in syseq:
    expr3 = eq
    print(
        solve([expr1.subs(expr3.lhs, expr3.rhs), 
               expr2.subs(expr3.lhs, expr3.rhs)], x),
        solve(expr3, n))
```
